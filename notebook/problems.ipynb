### Exercise 1

We will use NumPy to obtain information to describe statistically.

- Generate an array of 100 elements following a normal distribution.
- Generate an array of 100 elements following a chi-square distribution with 3 degrees of freedom.
- Calculate the main metrics and statistical measures that best describe the two vectors.

````python
# TODO
import numpy as np
import scipy.stats as stats

array_1 = np.random.normal(size=100)
print(f"Array de distribución normal de 100 elementos: {array_1}")

array_2 = np.random.chisquare(3,100)
print(f"Array chi-square de 100 elemetos:  {array_2}")


#Mean
print(f"Media de array de normal distribución: {np.mean(array_1)}")
print(f"Media de array chisquare: {np.mean(array_2)}")


# Desviación estandar y varianza
print(f"Desviación de array de normal distribución: {np.std(array_1)}")
print(f"Desviación estandar de array chisquare: {np.std(array_2)}")
print(f"Varianza de array de normal distribución: {np.var(array_1)}")
print(f"Varianza de array de chisquare: {np.var(array_2)})")


# Moda
print(f"Moda de array de normal distribución: {stats.mode(array_1)}")
print(f"Moda de array chisquare: {stats.mode(array_2)}")


# Mediana
print(f"Mediana de array de normal distribución: {np.median(array_1)}")
print(f"Mediana de array chisquare: {np.median(array_2)}")


#Rango
rango_1 = max(array_1) - min(array_1)
print(f"Rango de array de normal distribución: {rango_1}")
rango_2 = max(array_2) - min(array_2)
print(f"Rango de array de chisquare: {rango_2}")


#Asimetría y kurtosis
from scipy.stats import skew

print(f"Asimetría de array de normal distribución: {skew(array_1)}")
print(f"Asimetría de array de chisquare: {skew(array_2)}")

from scipy.stats import kurtosis

print(f"Kurtosis de array de normal distribución: {kurtosis(array_1)}")
print(f"Kurtosis de array chisquare: {kurtosis(array_2)}")
````

### Exercise 2

Write a Python program to calculate the standard deviation of the following data:

```py
data = [4, 2, 5, 8, 6]
```

````python
# TODO
import numpy as np
import math


data = [4, 2, 5, 8, 6]
n = len(data)

def d_standar(data):
  

    #1 la media
    media = np.mean(data)

   
    #2 restar la media a cada dato y elevar al cuadrado
   
    desviaciones_cuadradas = [(x - media)**2 for x in data]
   
    '''for x in data:
        resta = x - media
        cuadrado = resta ** 2'''
       
        

    #3 Sumar las diferencias cuadradas
    suma_cuadrados = np.sum(desviaciones_cuadradas) 
    
    #4 Calculamos la varianza
    varianza = suma_cuadrados / (n - 1)

    #5 Aplicar la raíz cuadrada
    raiz_cuadrada = math.sqrt(varianza)
    return raiz_cuadrada



d_standar([data])
print(f'la desviación estandar de data es: {d_standar(data)}')
````
